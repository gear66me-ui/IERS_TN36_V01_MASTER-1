```python
# %load IERS_0001_ECLIPTICAL_PLANE_ANALYSIS.py
"""
IERS_0001_ECLIPTICAL_PLANE_ANALYSIS.py
Part A
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

AU_KM = 149597870.7
ARCSEC_PER_RAD = 206264.80624709636

OBSERVER = {
    "name": "Bucaramanga",
    "latitude_deg": 7.12539,
    "longitude_deg": -73.1198,
    "elevation_m": 959.0,
}

TRANSIT = {
    "target": "Venus",
    "center": "Sun",
    "date": "2012-06-06",
}

def normalize(v):
    v = np.asarray(v, dtype=float)
    n = np.linalg.norm(v)
    if n == 0:
        raise ValueError("Zero-length vector")
    return v / n

def project_to_plane(vec, los):
    los = normalize(los)
    return vec - np.dot(vec, los) * los

def angular_separation(v1, v2):
    a = normalize(v1)
    b = normalize(v2)
    c = np.clip(np.dot(a, b), -1.0, 1.0)
    return np.arccos(c) * ARCSEC_PER_RAD

def linear_track_fit(x, y):
    return np.polyfit(x, y, 1)

def evaluate_track(coeff, x):
    return np.polyval(coeff, x)

def main():
    print("IERS TN36 - Ecliptical Plane Analysis")
    print("Part A initialized")
    print("Observer:", OBSERVER["name"])
    print("Transit:", TRANSIT["date"])
    print("Awaiting JPL Horizons vectors...")

if __name__ == "__main__":
    main()
```

```python
import os
import requests
import pandas as pd
from IPython.display import display

def human_size(size_bytes):
    size = float(size_bytes)
    for unit in ["B", "KB", "MB", "GB"]:
        if size < 1024 or unit == "GB":
            return f"{size:,.2f} {unit}"
        size /= 1024

# ----------------------------------------------------------
# 1. LARGEST FILES CURRENTLY IN COLAB
# ----------------------------------------------------------

local_files = []

for root, dirs, files in os.walk("/content"):
    if root.startswith("/content/sample_data"):
        continue

    for filename in files:
        path = os.path.join(root, filename)

        try:
            size = os.path.getsize(path)
            local_files.append({
                "Size": human_size(size),
                "Bytes": size,
                "File": path
            })
        except OSError:
            pass

local_df = pd.DataFrame(local_files)

print("LARGEST FILES IN COLAB")
if len(local_df):
    display(
        local_df
        .sort_values("Bytes", ascending=False)
        .head(30)[["Size", "File"]]
        .reset_index(drop=True)
    )
else:
    print("No files found.")

# ----------------------------------------------------------
# 2. LARGEST FILES CURRENTLY IN THE GITHUB REPOSITORY
# ----------------------------------------------------------

owner = "gear66me-ui"
repo = "IERS_TN36_V01_MASTER-1"
branch = "main"

api_url = (
    f"https://api.github.com/repos/{owner}/{repo}"
    f"/git/trees/{branch}?recursive=1"
)

response = requests.get(api_url, timeout=30)
response.raise_for_status()
tree = response.json()["tree"]

github_files = []

for item in tree:
    if item.get("type") == "blob":
        size = item.get("size", 0)

        github_files.append({
            "Size": human_size(size),
            "Bytes": size,
            "File": item["path"]
        })

github_df = pd.DataFrame(github_files)

print("\nLARGEST FILES IN GITHUB")
display(
    github_df
    .sort_values("Bytes", ascending=False)
    .head(30)[["Size", "File"]]
    .reset_index(drop=True)
)
```
